# Mapas de Color

```python
import sys
import matplotlib
# matplotlib.use("Agg")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(50)
```

```python
data = np.random.random((8,8))
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data)

cbar = ax.figure.colorbar(im, ax = ax)
cbar.ax.set_xlabel("Semana", rotation = 0, va = "top")
cbar.ax.set_ylabel("Producción %", rotation = -90, va = "bottom")

plt.show()
```
