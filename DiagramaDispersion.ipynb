```python
import sys
import matplotlib
# matplotlib.use("Agg")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(123)
```

```python
x1 = np.random.randint(0, 200, 100)
y1 = np.random.randint(0, 200, 100)

x2 = np.random.randint(50, 250, 120)
y2 = np.random.randint(50, 250, 120)
```

```python
plt.scatter(x1, y1, c = "green", label = "Group A")
plt.scatter(x2, y2, c = "salmon", label = "Group B")
plt.legend()
plt.show()
```
