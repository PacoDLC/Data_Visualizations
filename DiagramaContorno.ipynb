# Diagrama de Contorno

```python
import sys
import matplotlib
# matplotlib.use("Agg")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, Y = np.meshgrid(np.linspace(-12, 12, 180),
                   np.linspace(-20, 20, 200))
Z = np.sqrt(X ** 2 + Y ** 2)
```

```python
fig, ax = plt.subplots()
cnt = ax.contourf(X, Y, Z)

cbar = ax.figure.colorbar(cnt, ax = ax)
cbar.ax.set_ylabel("Temperatura", rotation = -90, va = "bottom")

plt.show()
```
