# Practica 1: Visualización de Datos

## Comparación de Rendimiento de Algoritmos

```python
import pandas as pd
```

```python
dataset_df = pd.read_csv("C:/Users/Admin/OneDrive/Curso Maestro Visualizaciones y Análisis de Datos en Python/data/comparasion_algoritmos.csv")
dataset_df.head(10)
```

```python
graph_1 = dataset_df.plot.bar(
    x = "dataset", 
    y = ["algoritmo1", "algoritmo2"], 
    title = "Comparación de Rendimiento de Algoritmos",
    width = 0.8,
    figsize = (12,5)
)
graph_1.set_xlabel("Datasets")
graph_1.set_ylabel("Rendimiento")
graph_1.grid(visible = True)
graph_1.legend(labels = ["Algoritmo 1", "Algoritmo 2"], loc = "upper left", bbox_to_anchor = (0,1))
graph_1.set_axisbelow(True)
```
