# Practica 2: Visualización de Datos

## Distribución de datos en Datasets

Importamos las librerías a emplear

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importamos e; conjunto de datos que vamos a emplear

```python
dataset_df = pd.read_csv("C:/Users/Admin/OneDrive/Curso Maestro Visualizaciones y Análisis de Datos en Python/data/datasets_top_4.csv", 
                         index_col = "dataset")
dataset_df
```

```python
dataset_df.info()
```

Dividimos el dataset en dos partes: parte 1

```python
subdataset_01 = dataset_df.iloc[:14,:]
subdataset_01
```

Dividimos el dataset en dos partes: parte 2

```python
subdataset_02 = dataset_df.iloc[14:,:]
subdataset_02
```

```python
Realizamos el gráfico con los subgraficos correspondientes
```

```python
graph_2, (sub_graph_11, sub_graph_12) = plt.subplots(1,2)
graph_2.set_size_inches(7,4)

subdataset_01.plot.barh(
    stacked = True,
    ax = sub_graph_11,
    width = 0.4,
    color = {"cat1": "magenta", "cat2": "purple", "cat3": "pink", "otros": "gray"}
)

subdataset_02.plot.barh(
    stacked = True,
    ax = sub_graph_12,
    width = 0.4,
    color = {"cat1": "magenta", "cat2": "purple", "cat3": "pink", "otros": "gray"}
)

sub_graph_11.set_xlabel("Porcentaje")
sub_graph_12.set_xlabel("Porcentaje")

sub_graph_11.set_ylabel("Dataset")
sub_graph_12.set_ylabel(" ")

sub_graph_11.legend().remove()

sub_graph_12.legend(
    ['Categoría 1', 'Categoría 2', 'Categoría 3', 'Otros'],
    ncol = 4,
    loc = "upper left",
    bbox_to_anchor = (-1.2,-0.15)
)

sub_graph_12.yaxis.tick_right()

sub_graph_11.text(70,-5,
    "Algunos datasets han sido ignorados del análisis",
    fontsize = 6,
    bbox = {"facecolor": "orange", "alpha": 0.5}
)

graph_2.suptitle("¿Cómo se distribuyen las categorías\ndentro de los Datasets?")
```
