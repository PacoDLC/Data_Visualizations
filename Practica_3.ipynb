## Anáilisis de Top N

Importamos las librerías a utilizar

```python
import pandas as pd
```

Importamos el conjunto de datos que vamos a utilizar para el análisis

```python
dataset_df = pd.read_csv("C:/Users/Admin/OneDrive/Curso Maestro Visualizaciones y Análisis de Datos en Python/data/viajes_surfistas.csv")
dataset_df
```

Identificamos los países (sin repetir) de los surfistas que participan

```python
dataset_df["homecountry"].unique()
```

Contamos cuántos países participan

```python
len(dataset_df["homecountry"].unique())
```

Para saber cuántos surfistas participan por país, realizamos una agrupación de datos

```python
agr_por_pais = dataset_df.groupby("homecountry").agg({"homename": 'count'})
agr_por_pais
```

Ordenamos de manera descendente para saber qué país tiene más surfistas

```python
agr_por_pais_ord = agr_por_pais.sort_values("homename", ascending = False)
agr_por_pais_ord
```

Identificamos el top 4 de países con más surfistas

```python
top4_df = agr_por_pais_ord.head(4)
top4_df
```
